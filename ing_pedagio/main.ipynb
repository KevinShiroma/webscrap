```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
from IPython.display import display

def extrair_varias_tabelas(url):
    # Fazendo a requisição HTTP para obter o conteúdo da página
    response = requests.get(url)

    # Verificando se a requisição foi bem-sucedida
    if response.status_code == 200:
        # Parseando o conteúdo da página com BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Encontrando a div com id 'parent-fieldname-text'
        div_content = soup.find('div', id='parent-fieldname-text')
        
        if div_content:
            # Extraindo todas as tags <strong> e as tabelas dentro da div
            strong_tags = div_content.find_all('strong')
            tables = div_content.find_all('table')
            
            # Criando um dicionário para armazenar DataFrames
            dataframes = {}
            
            # Iterando sobre as tabelas encontradas e transformando em DataFrames
            for idx, (strong, table) in enumerate(zip(strong_tags, tables)):
                table_html = str(table)
                df = pd.read_html(StringIO(table_html))[0]  # Transformando a tabela HTML em DataFrame
                
                # Criando um nome para o DataFrame
                name = f"DataFrame_{idx+1}"
                # Armazenando o DataFrame no dicionário
                dataframes[name] = df
                
                # Exibindo o conteúdo da tag <strong>
                print(f"\nConteúdo da tag <strong> {idx+1}:")
                print(strong.get_text(strip=True))
                
                # Exibindo o DataFrame
                print(f"\nDisplaying {name}:")
                display(df)
                
            return dataframes
        else:
            print("Div com o id 'parent-fieldname-text' não encontrada.")
            return {}
    else:
        print(f"Erro ao acessar a página: Status code {response.status_code}")
        return {}

# URL do exemplo
url_valores = "https://www.gov.br/antt/pt-br/assuntos/rodovias/concessionarias/lista-de-concessoes/contratos-encerrados/crt/tarifas-de-pedagio"

# Extraindo e exibindo os DataFrames
dataframes = extrair_varias_tabelas(url_valores)
```
